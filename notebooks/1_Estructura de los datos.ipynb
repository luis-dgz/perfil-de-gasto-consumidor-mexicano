# Encuesta ENIGH 2024

La Encuesta Nacional de Ingresos y Gastos de los Hogares (ENIGH) 2024 tiene como objetivo proporcionar un panorama estadístico del comportamiento de los ingresoos y gastos de los hogares mexicamos. La edición 2024 se realizó el levantamiento entre el 21 de agosto al 28 de noviembre de 2024, esta encuesta es realizada cada dos años.

Dentro de esta encuesta se levantan 5 cuestionarios de los cuales se recaba información acerca la vivienda seleccionada, las personas que intregan el hogar, ingresos del hogar, condición de ocupación de los integrantes del hogar y gastos del hogar, entre otros puntos. 

El objetivo principal es entender los perfiles de gasto por hogar mexicano, por lo cual nos enfocaremos principalmente en el **Cuestionario de gastos del hogar**. Este tiene como objetivo recabar información sobre todos los gastos que realizan los hogares. Los gastos se ordenan según tres periodos: mensuales, trimestrales y semestrales. Los periodos se determinaron según la frecuencia con la cual se compran los productos y el monto del gasto.

Fuente: [ENIGH 2024](https://www.inegi.org.mx/programas/enigh/nc/2024/#documentacion)
